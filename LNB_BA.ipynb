```python
import numpy as np
import statistics as stats
data = [12, 15, 12, 18, 16, 14, 19, 12]
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)
variance = np.var(data)
std_dev = np.std(data)
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.Series([10, 12, 15, 14, 13, 80, 16, 18, 19, 20])

plt.figure(figsize=(6, 4))
sns.boxplot(data=data)
plt.title('Outlier Detection using Boxplot')
plt.show()

sns.scatterplot(data=data)
plt.title('Outlier Detection using Scatterplot')
plt.show()
```

```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B', 'C', 'C', 'C', 'A', 'B'],
    'Value': np.random.randint(10, 100, 9)
})

# Random Sampling: Select 4 random samples from the data
random_sample = data.sample(n=4)

# Stratified Sampling: Sample 50% (frac=0.5) from each category
stratified_sample = data.groupby('Category', group_keys=False).apply(lambda x: x.sample(frac=0.5))

print("Random Sample:\n", random_sample)
print("\nStratified Sample:\n", stratified_sample)
```

```python
from scipy import stats
import numpy as np

# t-test for independent samples
group1 = [23, 21, 19, 24, 30]
group2 = [29, 31, 23, 34, 26]

t_stat, p_val = stats.ttest_ind(group1, group2)
print('T-test p-value:', p_val)

# Chi-square test for independence
obs = np.array([[10, 20],
                [20, 30]])

chi2, p, dof, exp = stats.chi2_contingency(obs)
print('Chi-square p-value:', p)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression().fit(x, y)
y_pred = model.predict(x)

plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.title('Simple Linear Regression')
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.DataFrame({
    'X1': [1, 2, 3, 4, 5],
    'X2': [2, 1, 4, 3, 5],
    'Y': [3, 3, 6, 5, 7]
})

X = data[['X1', 'X2']]
Y = data['Y']

model = LinearRegression().fit(X, Y)

print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
```
